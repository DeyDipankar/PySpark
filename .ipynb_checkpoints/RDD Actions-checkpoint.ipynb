```python
# working with files

airlines = sc.textFile('airlines.txt')
airlines.take(5)
```

```python
airlines.first() # returns the first element
```

```python
airlines.count() # returns total no. of elements
```

```python
airlines.distinct().count()
```

```python
for key,value in airlines.countByValue().items():
    if value > 3:
        print(key,value)
# it returns all names where those names appeared more than 5 times
```

```python
len(airlines.distinct().countByValue())
```

```python
# here we see that if we distinct, it returns only distinct entries. So, for value>1 doesn't return anything as
# each distinct entry is consered
for key,value in airlines.distinct().countByValue().items():
    if value > 1:
        print(key,value)
```

```python
#GLOM - retruns partitions as tuples

airlines = sc.textFile('airlines.txt', minPartitions = 1500) # we can define no.of partitions, default is 1

partition = airlines.glom().take(5) # taking first 5 partitions

for i in range(len(partition)):
    print(partition[i],len(partition[i]),'\n\n')
```
